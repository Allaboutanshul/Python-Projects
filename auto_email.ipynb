```python
import smtplib
from email.message import EmailMessage

msg = EmailMessage()
msg.set_content("This email was shared by python code.....") #
msg["Subject"] = "Hello For Python ......" #
msg["From"] = "anshulkumawat111@gmail.com"
msg["To"] = "southharmington@gmail.com" #

# REPLACE THIS with your 16-character code (NO SPACES)
APP_PASSWORD = "your16digitcode" 

try:
    with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server: #
        server.login("anshulkumawat111@gmail.com", APP_PASSWORD) #
        server.send_message(msg) #
    print("✅ Email sent successfully!")
except Exception as e:
    print(f"❌ Failed to send email. Error: {e}")
```
